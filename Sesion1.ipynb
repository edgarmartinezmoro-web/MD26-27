## Variables lógicas y tablas de verdad

```python
def negacion(p):  ### Negación
    return not p

print(f"{'p':<10}{'¬ p':<10}")
print("-" * 15) 
for p in [True, False]:
    a = negacion(p)
    print(f"{str(p):<10}{str(a):<10}")
```

```python
def conjuncion(p, q):  ### Conjunción
    return p and q

print(f"{'p':<10}{'q':<10}{'p ∨ q':<10}")
print("-" * 25) 
for p in [True, False]:
    for q in [True, False]:
        a = conjuncion(p, q)
        print(f"{str(p):<10}{str(q):<10}{str(a):<10}")
```

```python
def disyuncion(p, q):   ### Disyuncion
    return p or q

print(f"{'p':<10}{'q':<10}{'p ∧ q':<10}")
print("-" * 25) 
for p in [True, False]:
    for q in [True, False]:
        a = disyuncion(p, q)
        print(f"{str(p):<10}{str(q):<10}{str(a):<10}")
```

```python
def dix(p, q):      ### XOR
    return p ^ q

print(f"{'p':<10}{'q':<10}{'p xor q':<10}")
print("-" * 28) 
for p in [True, False]:
    for q in [True, False]:
        a = dix(p, q)
        print(f"{str(p):<10}{str(q):<10}{str(a):<10}")
```

 $$\neg \Big( P \rightarrow (Q \rightarrow \neg(P \rightarrow P)) \Big)$$

```python
def reconocer(p, q):      ### ¿qué es?
    return (not ( (not p) or ((not q) or (not ( not p or p)))))

print(f"{'p':<10}{'q':<10}{'p ?? q':<10}") 
print("-" * 28) 
for p in [True, False]:
    for q in [True, False]:
        a = reconocer(p, q)
        print(f"{str(p):<10}{str(q):<10}{str(a):<10}")
```

## Solución del problema inicial

```python
def resolver_acertijo():
    # Las opciones posibles para cada trol: True (Caballero) o False (Villano)
    opciones = [True, False]
    print(f"{'Trol 1':<10}{'Trol 2':<10}{'Trol 3':<10}")
    print("-" * 30)
    # Probamos todas las combinaciones posibles (las 8 combinaciones)
    for t1 in opciones:
        for t2 in opciones:
            for t3 in opciones:
                
                # Contamos cuántos caballeros hay en la combinación actual
                num_caballeros = sum([t1, t2, t3])
                
                # DECLARACIÓN TROL 1: "Si soy villano, entonces hay exactamente 2 caballeros"
                # En lógica, "Si A entonces B" es equivalente a "(no A) o B"
                # O visto de otra forma: si t1 es falso, la condición se cumple solo si hay 2 caballeros.
                dec_t1 = (not (not t1)) or (num_caballeros == 2)
                
                # DECLARACIÓN TROL 2: "El trol 1 está mintiendo"
                # t2 dice la verdad si y solo si la declaración de t1 es falsa
                dec_t2 = (dec_t1 == False)
                
                # DECLARACIÓN TROL 3: "O todos somos villanos, o al menos uno es caballero"
                todos_villanos = (t1 == False and t2 == False and t3 == False)
                al_menos_un_caballero = (num_caballeros >= 1)
                dec_t3 = todos_villanos or al_menos_un_caballero
                
                # REGLA DE ORO: Un trol dice la verdad SI Y SOLO SI es un Caballero.
                # Por lo tanto, el valor del trol (t) debe coincidir con el valor de su declaración (dec).
                if (t1 == dec_t1) and (t2 == dec_t2) and (t3 == dec_t3):
                    # Traducimos True/False a Caballero/Villano para que se entienda mejor
                    r1 = "Caballero" if t1 else "Villano"
                    r2 = "Caballero" if t2 else "Villano"
                    r3 = "Caballero" if t3 else "Villano"
                    
                    print(f"{r1:<10}{r2:<10}{r3:<10} <- ¡ESTA ES LA SOLUCIÓN!")
                    return

resolver_acertijo()
```
